```python
import pandas as pd 

import matplotlib.pyplot as plt 

import seaborn as sns 
```

```python
## Import dataset
score = pd.read_csv('/Users/Yedou/Desktop/SeabornScore/ign_scores.csv')
```

```python
score
```

```python
score.head()
```

```python
# Set Platform as index

score.set_index('Platform', inplace = True)

score.head()
```

```python
# Score of racing games on different platforms

# The size of our  bar plot

plt.figure(figsize=(12,7))


sns.barplot(x = score['Racing'], y = score.index)

plt.xlabel('Racing')

plt.ylabel('Platform')

plt.title('Score of racing games on different platforms')

plt.show()
```

```python
# Score of racing games on different platforms

# The size of our  bar plot

plt.figure(figsize=(12,7))


sns.barplot(x = score['Adventure'], y = score.index)

plt.xlabel('Adventure')

plt.ylabel('Platform')

plt.title('Score of Adventure games on different platforms')

plt.show()
```

```python
# heatmap 

plt.figure(figsize=(12,8))


sns.heatmap(data = score , annot= True)

plt.xlabel('Game')

plt.ylabel('Platform')

plt.show()
```
